```python
import urllib.request
import urllib.parse
import urllib.error
from bs4 import BeautifulSoup
import ssl
import json
from urllib.request import urlopen
from bs4 import BeautifulSoup 
import requests
import csv
```

```python
#For ignoring SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
url="https://www.amazon.com/Ring-Doorbell-Motion-Activated-Installation/product-reviews/B00N2ZDXW2/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews&pageNumber="
```

```python
product_json = {}
product_json['short-reviews'] = []
product_json['long-reviews'] = []
#statrt_url=[]
for i in range(1, 500):   #Number of pages plus one
    start_url=url+str(i)
    html = (urllib.request.urlopen(start_url, context=ctx).read())
    #htmlall.append(html)
    soup = BeautifulSoup(html, "html5lib")
    soup.prettify('utf-8')
    for a_tags in soup.findAll('a',
                               attrs={'class': 'a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold'
                               }):
        short_review = a_tags.text.strip()
        product_json['short-reviews'].append(short_review)
    # This block of code will help extract the long reviews of the product
    for spans in soup.findAll('span', attrs={'data-hook': 'review-body'
                             }):
        long_review = spans.text.strip()
        product_json['long-reviews'].append(long_review)
```
