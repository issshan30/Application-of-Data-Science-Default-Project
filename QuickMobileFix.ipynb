```python
from bs4 import BeautifulSoup
import math
import pandas as pd
import requests
#Dataframe
final_list=[]
url = 'https://uk.trustpilot.com/review/www.quickmobilefix.com'
#making a request to get the url data
r=requests.get(url)
soup = BeautifulSoup(r.text, 'lxml')
review_count_h2=soup.find('h2',class_="header--inline").text
review_count=int(review_count_h2.strip().split(' ')[0].replace(',', '').strip())
#there are 20 reviews per page so total pages can be calculated as
pages=int(math.ceil(review_count/20))
#Range to 1 to pages+1
for pg in range(1, pages+1):
  pg = url + '?page=' + str(pg)
  r=requests.get(pg)
  soup = BeautifulSoup(r.text, 'lxml')
  for paragraph in soup.find_all('section',class_='review__content'):
     try:
         title=paragraph.find('h2',class_='review-content__title').text.strip()
         content=paragraph.find('p',class_='review-content__text').text.strip()
         final_list.append([title,content])
     except AttributeError:
        pass
df = pd.DataFrame(final_list,columns=['Title','Content'])
print(df)
```
